```python
#Quiz-01
num = input("Enter a number: ")  

num = int(num)
if num % 2 == 0:
  print(num, "is even")
else:
  print(num, "is odd")
```

```python
#Quiz-02

radius = float(input("Enter the radius of the circle: "))
PIE_VALUE = 3.14

if radius <= 0:
  print("Radius cannot be negative or zero")
else:
  circumference = 2 * PIE_VALUE * radius
  print("The Circumference of a circle with radius", radius, "is", circumference)
```

```python
#Quiz-03

hight = float(input("Enter the hight of the triangle: "))
base = float(input("Enter the base of the triangle: "))

if hight <= 0 and base <= 0:
  print("Hight and base cannot be negative or zero")
else:
  hypotenuse = (hight**2 + base**2) ** 0.5
  print("The area of a triangle with hight", hight, "and base", base, "is", hypotenuse)
```
