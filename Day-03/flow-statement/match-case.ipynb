```python
x = 2

match x:
  case 1:
    print("x is 1")
  case 2:
    print("x is 2")
  case 3:
    print("x is 3")
  case _:
    print("x is something else")
```

```python
#Quiz-07
day = input("Enter day : ")

match day:
  case "monday":
    print("It is a week day")
  case "tuesday":
    print("It is a week day")
  case "wednesday":
    print("It is a week day")
  case "thursday":
    print("It is a week day")
  case "friday":
    print("It is a week day")
  case "saturday":
    print("It is a weekend day")
  case "sunday":
    print("It is a weekend day")
  case _:
    print("Invalid day")
```

```python
#Quiz-07
day = input("Enter day : ")

match day:
  case "monday"|"tuesday"|"wednesday"|"thursday"|"friday":
    print("It is a week day")
  case "saturday"|"sunday":
    print("It is a weekend day")
  case _:
    print("Invalid day")
```

```python
numbers = [4,3,7]

match numbers:
  case [x,y]:
    print(x * y)
  case [x,y,z]:
    print(x + y + z)
  case _:
    print("Invalid")
```

```python
#Quiz-08
number = [2,5,6]

match number:
  case [x,y]:
    hypotence = (x ** 2 + y ** 2) ** 0.5
    print("Hypotence is", hypotence)
  case [x,y,z]:
    semi_perimeter = (x + y + z) / 2
    area = (semi_perimeter * (semi_perimeter - x) * (semi_perimeter - y) * (semi_perimeter - z)) ** 0.5
    print("Area is", area)
  case _:
    print("Un Supported number of sides")
```
